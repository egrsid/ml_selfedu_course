```python
import numpy as np

x_test = np.array([(5, -3), (-3, 8), (3, 6), (0, 0), (5, 3), (-3, -1), (-3, 3)])

# 
X_test = np.concatenate([np.ones(x_test.shape[0]).reshape(-1, 1), x_test], axis=1)  #  1   ,       
w = [-33/13, 9/13, 1]
W = np.array(w).reshape(1, -1)
predict = [int(xx) for xx in np.sign(W@X_test.T)[0]]
predict
```

```python
import numpy as np

x_test = [(9, 6), (2, 4), (-3, -1), (3, -2), (-3, 6), (7, -3), (6, 2)]

# 
w = np.array([1, -1/2, 5/14])
X_test = np.array([[1, x1, x2] for x1, x2 in x_test])
predict = np.sign(w.reshape(1, -1) @ X_test.T).squeeze()
predict
```
